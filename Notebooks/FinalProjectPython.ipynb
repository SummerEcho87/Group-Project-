```python
#examine genres, lists are in quotes so need to be converted from srting to list
netflix_df['genres'].unique()
```

```python
netflix_df["genres"] = netflix_df["genres"].apply(lambda s: eval(s))
```

```python
#check if it worked
netflix_df.loc[1,'genres']
```

```python
#create set to see how many unique genres
genres_unique = set()
```

```python
#go through each row and add unique genres to set
for l in netflix_df.genres.values:
    genres_unique = genres_unique.union(set(l))
genres_unique = list(genres_unique)
```

```python
netflix_recode_df = netflix_df.copy()
```

```python
#create columns for each unique genre, 99 is used as a conventional place holder
netflix_recode_df[genres_unique] = [99]*len(genres_unique)
```

```python
#dummy code each row
for col in genres_unique:
    for i in range(len(netflix_recode_df)):
        if col in netflix_recode_df.loc[i,'genres']:
            netflix_recode_df.loc[i,col] = 1
        else:
            netflix_recode_df.loc[i,col] = 0 
```

```python
#drop original genre column
netflix_recode_df = netflix_recode_df.drop('genres',axis=1)
```

```python
#convert countries from string to list
netflix_recode_df["production_countries"] = netflix_recode_df["production_countries"].apply(lambda s: eval(s))
```

```python
#see how many unique countries
production_countries_unique = set()
```

```python
for l in netflix_recode_df.production_countries.values:
    production_countries_unique = production_countries_unique.union(set(l))
production_countries_unique = list(production_countries_unique)
```

```python
production_countries_unique = list(production_countries_unique)
```

```python
len(production_countries_unique)
#109 countries where content was produced
```

```python
#see how many movies were produced in each country
country_dict = {}
```

```python
for l in netflix_recode_df.production_countries.values:
    for country in l:
        if country_dict.get(country) == None:
            country_dict[country]=1
        else:
            country_dict[country] += 1
```

```python
#sort dictionary by values
country_dict = {k: v for k, v in sorted(country_dict.items(), key=lambda item: item[1], reverse=True)}
```
