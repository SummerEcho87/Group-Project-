# Perform a Kruskall-Wallis Test to test for analysis of variance when non-parametic (in our case uneven sample sizes)

**Kruskall-Wallis Test for variance of runtime of shows by year**

```r
kruskal.test(runtime ~ Realease_Year1, data = NetflixRSHOW) 
```

**P-value shows a significant difference in the average show runtime**

**Kruskall-Wallis Test for variance of runtime of movies by year**

```r
kruskal.test(runtime ~ Realease_Year1, data = NetflixRMOVIE) 
```

**Since this p-value is less than 0.05, we can reject the null hypothesis that the median runtime is the same across the years**

# T-TEST for Movies - Is there a significant difference between the average Netflix Movie IMDB score and the overall IMDB score for streaming services?

```r
t_obj <- t.test(NetflixMOVIE$imdb_score, mu = 6.4)
print(t_obj)
```

**P-value is less than .05. We can see that Netflix Movies have a significantly lower IMDB score at 6.2 than the overall average score of 6.4**

# T-TEST for Shows - Is there a significant difference between the average Netflix Show IMDB score and the overall IMDB score for streaming services?

```r
t_obj <- t.test(NetflixSHOW$imdb_score, mu = 7.0)
print(t_obj)
```

**P-value is greater than .05 so there is NOT a signifcant difference in averages. The Average Netflix score of 7.0 is the same as the overall average of 7.0**
